```python
import pandas as pd
import numpy as py
```

```python
df = pd.read_csv("/content/Data-cleaning-for-beginners-using-pandas.csv")
```

```python
df
```

```python
print(df.isnull())
```

```python
df['Index'].fillna(df['Index'].mean(), inplace=True)
```

```python
print(df.isnull())
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
print(df.isnull())
```

```python
df['Rating'].fillna(df['Rating'].mean(), inplace=True)
print(df.isnull())
```

```python
df
```

```python
print(df.dtypes)
```

```python
df['Age'] = df['Age'].astype(int)
print(df.dtypes)
df
```

```python
df['Rating'] = df['Rating'].round(2)
df
```

```python
df['Location'] = df['Location'].astype(str)
df['Easy Apply'] = df['Easy Apply'].astype(str)
df
```

```python
print(df.dtypes)
```

```python
df['Location'] = df['Location'].replace(["India,In","Australia Aus","New York,Ny","India In"],["India","Australia","New York","India"])
df
```

```python
df['Established'] = df['Established'].replace(-1,0)
df['Easy Apply'] = df['Easy Apply'].replace("-1","0")
df
```

```python
df['Established'] = df['Established'].replace(0,df['Established'].mean()).round()
df['Established'] = df['Established'].astype(int)
df['Easy Apply'] = df['Easy Apply'].replace("TRUE",1).astype(int)
df['Salary'] = df['Salary'].replace("$"," ")
df
```
